<h4>7.2.3 原子指令</h4>
<h5>7.2.3.3 原子操作成本</h5>
<p>当全局或共享内存执行原子操作，必须把实际需要的值写入到全局或共享内存中</p>
<p>共享地址冲突的原子访问可能要求发生冲突的线程不断重试</p>
<p>线程在同一个线程束执行不同指令时，执行是序列化的</p>
<p>只在保证正确时使用不安全访问</p>
<h5>7.2.3.4 限制原子操作的性能成本</h5>
<p>使用局部操作来增强全局原子操作</p>
<h5>7.2.3.5 原子级浮点支持</h5>
<p>只有atomicExch/atomicAdd 支持单精度</p>
