<h2>6 流和并发</h2>
<h3>6.1 流和事件概述</h3>
<h4>6.1.1 CUDA流</h4>
<p>隐式声明的流（空流） 显式声明的流（非空流）</p>
<p>未创建默认使用空流 非空流可以被显式地创建和管理，重叠不同的CUDA操作必须使用非空流</p>
<p>基于流的异步内核启动和数据传输支持以下粗粒度并发：</p>
<p>重叠主机计算和设备计算 主机与设备间的数据传输 设备间的数据传输</p>
<p>当执行异步数据传输时，必须使用固定主机内存</p>
<h4>6.1.2 流调度</h4>
<h5>6.1.2.1 虚假的依赖关系</h5>
<p>即使多达16个网格同时执行，但所有的流最终被多路复用到单一的硬件工作队列中。虚假依赖关系指C                   依赖B， P在B-C流后，则只有P和C能并行，看似P依赖于C，但其实不然</p>
<h5>6.1.2.2 Hyper-Q技术</h5>
<p>使用多个硬件工作队列减少虚假依赖</p>
<h4>6.1.3 流的优先级</h4>
<p>cudaStreamCreateWithPriority 指定流优先级</p>
<p>cudaDeviceGetStreamPriorityRange</p>
<h4>6.1.4 CUDA事件</h4>
<p>CUDA事件本质是CUDA流中的标记，与该流内操作流中特顶点相关联 任务：同步流的执行 监控设备的进展 当一个给定CUDA流中先前所有操作都执行结束后，才完成</p>
<h4>6.1.5 流同步</h4>
<p>CUDA操作分为两大类 内存相关操作 内核启动</p>
<p>异步流（非空流）-> 阻塞流 非阻塞流 /同步流（空流）</p>
<h5>6.1.5.1 阻塞流和非阻塞流</h5>
<p>cudaStreamCreate创建阻塞流 流中操作可以被阻塞 空流为隐式流 和其他阻塞流同步</p>
<h5>6.1.5.2 隐式同步</h5>
<p>显式：cudaDeviceSynchronize...</p>
<p>隐式：cudaMemcpy隐式同步设备和主机</p>
<p>锁页主机内存分配 设备内存分配 设备内存初始化 统一设备两地址内存复制 一级缓存/共享内存配置修改</p>
<h5>6.1.5.3 显式同步</h5>
<p>同步设备 同步流 同步流中事件 使用事件跨流同步</p>

<h4>6.2.2 Fermi GPU上的虚假依赖关系</h4>
<p>由主机调度内核的顺序引起，由于所有流被多路复用到一个硬件队列中，所以前面的流阻塞了后面的流</p>
<h4>6.2.5 GPU资源的并发限制</h5>
<p>内核启动时可能会创建多个线程，可用的硬件资源会成为限制并发的主要因素，因为阻止启动符合条件内核</p>
<h4>6.2.6 默认流的阻塞行为</h4>
<p>默认流阻塞非空流， 直到默认流操作完成</p>

<h3>6.3 重叠内核执行和数据传输</h3>
<p>若内核使用数据A，那么对A数据传输必须在内核启动前。必须位于相同的流中</p>
<p>若内核完全不使用数据A，那么内核执行和数据传输可位于不同的流</p>
<h3>6.4 重叠GPU和CPU执行</h3>
<p>所有内核启动默认为异步，启动内核有效操作就可以自动重叠GPU/CPU</p>
<h3>6.5 流回调</h3>
<p>一旦流回调之前所有流操作全部完成，被流回调指定的主机端函数会被CUDAruntime调用</p>
